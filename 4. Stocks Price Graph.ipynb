```python
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
stocks = ['GOOG', 'AMZN','AAPL','MSFT','FB']
data = pdr.get_data_yahoo(stocks, start='2021-01-01', end='2022-02-03')['Close']
```

```python
data.head()
```

```python
rcParams['figure.figsize'] = 12,6
plt.plot(data.GOOG, linewidth = '2.5')
plt.plot(data.AMZN, linewidth = '2.5')
plt.plot(data.AAPL, linewidth = '2.5')
plt.plot(data.MSFT, linewidth = '2.5')
plt.plot(data.FB, linewidth = '2.5')
plt.grid(True, color='b', linestyle='dotted')
plt.title('The Big Five (Stock Prices)', fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price in USD", fontsize=12)
plt.style.use('seaborn-white')
plt.legend(['GOOGLE','AMAZON','APPLE','MICROSOFT','META (FB)'])
plt.show()
```
